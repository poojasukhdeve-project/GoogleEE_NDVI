```python
#import Libraries
import google.colab
import ee
import geemap
```

```python
# Authenticate and intialize
ee.Authenticate()
ee.Initialize()
```

```python
from jinja2.environment import reduce
# Map = geemap.Map(center = [23.8103,90.4125], zoom = 9)
# Map = geemap.Map(center = [19.7515,75.7139], zoom = 9)
#Map = geemap.Map()
#Map
Map = geemap.Map()
dataset1 = ee.FeatureCollection("USDOS/LSIB_SIMPLE/2017")

ndvi_Dataset = ee.ImageCollection("MODIS/061/MOD13Q1")
country = dataset1.filter(ee.Filter.eq('country_na', 'India'))

NDVI = ndvi_Dataset.select('NDVI')\
      .filterDate('2022-01-01', '2022-10-30')\
      .reduce(ee.Reducer.mean())\
      .clip(country)

ndviVis = {
  'min': 0.0,
  'max': 8000.0,
  'palette': [
    'FFFFFF', 'CE7E45', 'DF923D', 'F1B555', 'FCD163', '99B718', '74A901',
    '66A000', '529400', '3E8601', '207401', '056201', '004C00', '023B01',
    '012E01', '011D01', '011301'
  ],
};



Map.addLayer(country)
Map.addLayer(NDVI, ndviVis)
Map.centerObject(country, 13)
Map
```
